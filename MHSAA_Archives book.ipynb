```python
## Book to collect all of the archives of the MHSAA about baseball pre-2011

# I have a good stanard dataset from 2010-2011 season onward.

# This book is made to collect all the resources from the MHSAA website for baseball before that



# URL of the MHSAA results archive

archive_url = 'https://my.mhsaa.com/Sports/Baseball/Tournament-Results-Archive'


```

```python
import requests
import pandas as pd
import bs4 as BeautifulSoup
import re


```

```python
## Use beautiful soup and other tools to collect any and all data linked 
## directly from the archive page

# some of it wil lead to html documents with contest data, some will just be pdfs

# I want to collect all of it

## read the page

archive_page = requests.get(archive_url)

## parse the page

archive_soup = BeautifulSoup.BeautifulSoup(archive_page.text, 'html.parser')

## find all the links

archive_links = archive_soup.find_all('a')


## make a list of all the links

archive_links_list = []

## get all of the baseball links from page
for link in archive_links:
    archive_links_list.append(link.get('href'))


```
