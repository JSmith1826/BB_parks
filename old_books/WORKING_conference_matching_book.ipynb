```python
### Clean up the conference data from the file that I manually checked

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read in the data
df = pd.read_csv('TEMP/temp_cleaned_conference.csv')

df.info()
```

```python
### drop rows that have duplicates in the school name

df.drop_duplicates(subset='School', keep='first', inplace=True)

df.info()

## replace any instances of the words 'High School' with 'HS'

df['School'] = df['School'].str.replace('High School', '')

## Remove any numbers from the school name

df['School'] = df['School'].str.replace('\d+', '')

# remove and brackets from the school name

df['School'] = df['School'].str.replace('[', '')
df['School'] = df['School'].str.replace(']', '')

# # Drop any rows that contain the word 'Toledo'

# df = df[~df['School'].str.contains('Toledo')]

### Add HS to the end of the school name

df['School'] = df['School'] + ' HS'










df.head(50)
```

```python
df.info()

## Convert the school and conference names to strings

df['School'] = df['School'].astype(str)
df['conference'] = df['conference'].astype(str)

df.info()
```

```python
### Match the school name to the school name in the other data set and pull the school ID

# Read in the data

df2 = pd.read_csv('data/tables/mhsaa_enrolment_2022.csv')

df2.info()
```

```python
### Match the school name to the school name in the other data set and pull the school ID

# Read in the data

df2 = pd.read_csv('data/tables/mhsaa_enrolment_2022.csv')

## Import dependencies

import pandas as pd
import numpy as np
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

hs_names = df['School'].tolist()
mhsaa_names = df2['school_name'].tolist()

# set the treshold for the fuzzy match
treshold = 90
mat1 = []


## do the fuzzy match
for i in hs_names:
    mat1.append(process.extract(i, mhsaa_names, limit=1, scorer=fuzz.token_set_ratio))

df['match'] = mat1


### Take the first item in the tuple
df['match'] = df['match'].apply(lambda x: x[0][0])

#### Lookup the hs_df['match'] in the mhsaa_df['school_name'] and return the school_id
df['school_id'] = df['match'].apply(lambda x: df2[df2['school_name'] == x]['school_id'].values[0])
```

```python
df.head()

## DID NOT CHECK MACHES BEFORE OUTPUTTING FILE
## Drop the match column
df.drop(columns=['match'], inplace=True)

## Output the file
df.to_csv('data/tables/mhsaa_conference.csv', index=False)

```
