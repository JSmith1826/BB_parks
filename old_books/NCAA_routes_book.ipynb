```python
## get all color info in hex codes

## make sure color1 and color2 are all lower case
address_df['color1'] = address_df['color1'].str.lower()
address_df['color2'] = address_df['color2'].str.lower()

df = address_df

import pandas as pd
import webcolors

def convert_to_hex(color):
    # Remove leading '#'
    if color.startswith('#'):
        color = color[1:]
    # Try to convert color names to hex
    try:
        return webcolors.name_to_hex(color).upper()
    # If color is already hex code
    except ValueError:
        if len(color) == 6:
            try:
                int(color, 16)  # check if hex color code is valid
                return '#' + color.upper()
            except ValueError:
                pass
    return None

# Assuming df is your dataframe and 'color1' is the name of the column
df['color1'] = df['color1'].apply(convert_to_hex)
df['color2'] = df['color2'].apply(convert_to_hex)


```

```python
## Use google maps api to create lines (representing the quickest driving route) in the team colors from the visiting team to the host team

import googlemaps
from datetime import datetime

# Replace with your actual Google Maps API key
gmaps = googlemaps.Client(key='AIzaSyA_BhlTupRdBPBhRptQuR6pYorMVYQnRMA')

# Create a list of the polylines for each team
polylines = []

for i in range(len(df)):
    # Request directions via car
    now = datetime.now()
    origin = df['Origin'][i]
    end_point = df['End_point'][i]

    # If end_point is geographic coordinates in string format (lng, lat), convert it to tuple
    if isinstance(end_point, str) and ',' in end_point:
        # Strip brackets and spaces
        end_point = end_point.strip("[] ").split(',')
        # Swap latitude and longitude and convert to float
        end_point = (float(end_point[1]), float(end_point[0]))

    try:
        directions_result = gmaps.directions(origin, end_point, mode="driving", departure_time=now)
        polylines.append(directions_result[0]['legs'][0]['steps'][0]['polyline']['points'])
    except Exception as e:
        print(f"Failed to get directions for Origin: {origin}, End_point: {end_point}. Error: {e}")
        polylines.append(None)  # Add a None value if error occurs

## Add the polylines to the dataframe
df['polyline'] = polylines

df.head()

```
