```python
import os
import pandas as pd

# Directory with the csv files
dir = 'TEMP/clean_tables/year_by_year_summary/'

# List files and create a list of the files
files = os.listdir(dir)

# Create an empty list to store the results
results = []

# Loop through the files
for file in files:
    # Read the file into a dataframe
    df = pd.read_csv(dir + file)
    


    total_teams = df['teamName'].nunique()
    total_games = df['regular_season_games_played'].sum()
    avg_games = df['regular_season_games_played'].mean()
    total_playoff_games = df['playoff_games_played'].sum()
    avg_playoff_games = df['playoff_games_played'].mean()

    # Number of teams with more than 20 games played
    df_20 = len(df[df['regular_season_games_played'] > 15])

    # Append the results to the list
    results.append([file, total_teams, total_games, avg_games, total_playoff_games, avg_playoff_games, df_20])

# Create a dataframe from the list of results
df_results = pd.DataFrame(results, columns=['year_file','Total Lines',
                                            'Total Games', 
                                            'Avg Games per', 'Total Playoff Games', 
                                            'Average Playoff Games', "15+ Teams"])

df_results.head(20)



```

```python
## Get the playoff wins, loss, run_scored and run_allowed for each team in each file and 
## create a new dataframe

# Create an empty list to store the results
results = []

# Loop through the files
for file in files:
    # Read the file into a dataframe
    df = pd.read_csv(dir + file)
    
    # Create a dataframe with the playoff results
    df_playoffs = df[['teamName', 'playoff_games_played', 'playoff_W', 'playoff_L', 
                      'playoff_runs_scored_total', 'playoff_runs_allowed_total']]
    
    # Append the results to the list
    results.append(df_playoffs)

# group by teamName to get the total playoff wins, losses, runs scored and runs allowed
df_playoffs = pd.concat(results).groupby('teamName').sum()

# Create a new column with the playoff win percentage
df_playoffs['playoff_win_percentage'] = df_playoffs['playoff_W'] / df_playoffs['playoff_games_played']

# Create a new column with the playoff run differential
df_playoffs['playoff_run_differential'] = df_playoffs['playoff_runs_scored_total'] - df_playoffs['playoff_runs_allowed_total']

# Avergae differential
df_playoffs['playoff_avg_run_differential'] = df_playoffs['playoff_run_differential'] / df_playoffs['playoff_games_played']

# average wins per year
df_playoffs['playoff_avg_wins_per_year'] = df_playoffs['playoff_W'] / 10


## Display the results

# Sort the values by the playoff win percentage
df_playoffs.sort_values('playoff_win_percentage', ascending=False).head(20)




```

```python
### Collect the regular season data

# Create an empty list to store the results
results = []

# Loop through the files
for file in files:
    # Read the file into a dataframe
    df = pd.read_csv(dir + file)
    
    # Create a dataframe with the playoff results
    df_regular = df[['teamName', 'regular_season_games_played', 'regular_season_W', 'regular_season_L', 
                     'regular_season_T', 'regular_season_runs_scored_total', 'regular_season_runs_allowed_total']]
    
    # Append the results to the list
    results.append(df_regular)
```

```python
## Get the regular season wins, loss, tie run_scored and run_allowed for each team in each file and 
## create a new dataframe

# group by teamName to get the total regular season wins, losses, runs scored and runs allowed

df_regular = pd.concat(results).groupby('teamName').sum()

# Create a new column with the regular season win percentage
df_regular['regular_season_win_percentage'] = df_regular['regular_season_W'] / df_regular['regular_season_games_played']

# Create a new column with the regular season run differential
df_regular['regular_season_run_differential'] = df_regular['regular_season_runs_scored_total'] - df_regular['regular_season_runs_allowed_total']

# Avergae differential
df_regular['regular_season_avg_run_differential'] = df_regular['regular_season_run_differential'] / df_regular['regular_season_games_played']

# average wins per year
df_regular['regular_season_avg_wins_per_year'] = df_regular['regular_season_W'] / 10

## Display the results

# Sort the values by the regular season win percentage
df_regular.sort_values('regular_season_win_percentage', ascending=False).head(20)

## Filter results ot only teams that have avg wins per year > 10

df_regular[df_regular['regular_season_avg_wins_per_year'] > 10].sort_values('regular_season_win_percentage', ascending=False).head(20)


# sort by avg wins per year
df_regular.sort_values('regular_season_avg_wins_per_year', ascending=False).head(20)


```
