```python
###### NOtebook to calculate some statistics from the ballfield json file

## Dependencies
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Path to Ballpark JSON file
json_path = 'data\default_updated_output.json'

with open(json_path) as json_file:
    data = json.load(json_file)

# print(data)


# If the JSON data is a list of dictionaries, you can convert it to a pandas DataFrame
if isinstance(data, list) and all(isinstance(item, dict) for item in data):
    df = pd.DataFrame(data)

    # Display the first few rows to get a sense of the data
    display(df.head())

    # Get a summary of the DataFrame
    display(df.describe(include='all'))

    # Get data types of each column
    display(df.dtypes)
else:
    print("JSON data is not a list of dictionaries. Please provide more information about the data structure.")
```

```python
#### GET TOTAL FIELD COUNT AND COUNT BY LEVEL

## Print total fields
print("Total Fields: " + str(len(df)))

## Count of field by Levels
print('Fields by level:') 
print(df['level'].value_counts())
```

```python
## AREA CALCULATIONS

## Total area of the fop column
## Calculate the areas in square miles
fop_miles = df['fop_area_sqft'].sum() / 27878400
foul_miles = df['foul_area_sqft'].sum() / 27878400
total_miles = fop_miles + foul_miles

## Calculate the areas in acres
fop_acres = df['fop_area_sqft'].sum() / 43560
foul_acres = df['foul_area_sqft'].sum() / 43560
total_acres = fop_acres + foul_acres


print("Total Fair Territory: {:.3f} square miles".format(fop_miles))
print("Total Foul Territory: {:.3f} square miles".format(foul_miles))
print("Total Territory: {:.3f} square miles".format(total_miles))
print("----------------------------------------------")
print("Total Fair Territory: {:.3f} acres".format(fop_acres))
print("Total Foul Territory: {:.3f} acres".format(foul_acres))
print("Total Territory: {:.3f} acres".format(total_acres))

```

```python
### Figure out how long the boundry lines for for the fop and foul territory

import pandas as pd
from geopy.distance import distance

def calculate_boundary_distance(coordinates):
    total_distance = 0
    for i in range(len(coordinates) - 1):
        p1 = coordinates[i]
        p2 = coordinates[i + 1]

        # Calculate the distance between the two points in meters
        distance_meters = distance(p1[::-1], p2[::-1]).meters

        # Convert meters to feet
        distance_feet = distance_meters * 3.28084

        total_distance += distance_feet

    return total_distance

# # Sample DataFrame
# data = {
#     'fop': [
#         [[-84.36583, 46.4813183], [-84.3645398, 46.4813156], [-84.3645371, 46.4816997], [-84.3645385, 46.4817256], [-84.3647839, 46.4820331], [-84.3647745, 46.4820377], [-84.3652251, 46.4822095], [-84.3658206, 46.4822132], [-84.36583, 46.4813183]],
#     ],
#     'foul': [
#         [[-84.36583, 46.4813183], [-84.3645398, 46.4813156], [-84.3645371, 46.4816997], [-84.3645385, 46.4817256], [-84.3647839, 46.4820331], [-84.3647745, 46.4820377], [-84.3652251, 46.4822095], [-84.3658206, 46.4822132], [-84.36583, 46.4813183]],
#     ],
# }

df = pd.DataFrame(data)

# Calculate the total distance for each row in the DataFrame
df['fop_distance'] = df['fop'].apply(calculate_boundary_distance)
df['foul_distance'] = df['foul'].apply(calculate_boundary_distance)

# Calculate the sum of the distances for each column
total_fop_distance = df['fop_distance'].sum()
total_foul_distance = df['foul_distance'].sum()

print("Total FOP distance: {} feet".format(total_fop_distance))
print("Total Foul distance: {} feet".format(total_foul_distance))

## Conver the total distance to miles
total_fop_distance_miles = total_fop_distance / 5280
total_foul_distance_miles = total_foul_distance / 5280

print("Total FOP distance: {:.3f} miles".format(total_fop_distance_miles))
print("Total Foul distance: {:.3f} miles".format(total_foul_distance_miles))
print("Total distance: {:.3f} miles".format(total_fop_distance_miles + total_foul_distance_miles))


```
