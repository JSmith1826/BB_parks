```python
# Create a function to procewss the data when given a list of files

def process_data(files):
    # Initialize an empty list to hold the parsed data
    parsed_data = []

    # Loop over the files
    for file in files:
        # Read the file into a DataFrame
        df = pd.read_csv(file)


################## This is where I left off  ######################


        total_games = len(df)
        games_with_score = len(df.dropna(subset=['teamScore', 'opponentScore']))
        games_without_score = total_games - games_with_score

        print(f'Total games: {total_games}\nGames with score: {games_with_score}\nGames without score: {games_without_score}\nProportion of games with score: {games_with_score / total_games}')

        grouped = df.groupby('teamName')

        summary_stats = grouped.agg({
            'teamScore': ['count', 'mean', 'median', 'sum'],
            'opponentScore': ['count', 'mean', 'median', 'sum']
        }).reset_index()

        def calculate_results(row):
            if row['teamScore'] > row['opponentScore']:
                return 'win'
            elif row['teamScore'] < row['opponentScore']:
                return 'loss'
            else:
                return 'tie'

        df['result'] = df.apply(calculate_results, axis=1)

        results = df.groupby(['teamName', 'result']).size().unstack(fill_value=0).reset_index()

        location_summary = pd.crosstab(df['teamName'], df['homeOrAway'])

        tournament_games = df[df['contestType'].notnull()]

        tournament_summary = tournament_games.groupby('teamName').agg({
            'teamScore': ['count', 'mean', 'median', 'sum'],
            'opponentScore': ['count', 'mean', 'median', 'sum']
        }).reset_index()

        results['total_games'] = results['win'] + results['loss'] + results['tie']
        results['win_percentage'] = results['win'] / results['total_games']

        scheduled_vs_recorded = df.groupby('teamName').agg({
            'teamScore': ['count'],
            'gameDate': ['count']
        }).reset_index()

        scheduled_vs_recorded['scheduled_vs_recorded'] = scheduled_vs_recorded['teamScore']['count'] / scheduled_vs_recorded['gameDate']['count']

        scheduled_vs_recorded = scheduled_vs_recorded[['teamName', 'scheduled_vs_recorded']]

        summary_stats = pd.merge(summary_stats, results, on='teamName')

        summary_stats = pd.merge(summary_stats, location_summary, on='teamName')

        summary_stats = pd.merge(summary_stats, tournament_summary, on='teamName')

        summary_stats = pd.merge(summary_stats, scheduled_vs_recorded, on='teamName')

        ## Add a column for the year
        summary_stats['year'] = file[-8:-4]

        ## save the summary_stats dataframe to a csv file in their own folder
        summary_stats.to_csv(f'.sum_v1_{file[-8:-4]}.csv', index=False)

        # close function

    return parsed_data

# Call the function
process_data(files)

# # Create a function to process the data when given a list of files
# def process_data(files):
#     # Initialize an empty list to hold the parsed data
#     parsed_data = []

#     # Loop over the files
#     for file in files:
#         # Read the file into a DataFrame
#         df = pd.read_csv(file)

#         # Get the total number of games
#         total_games = len(df)

#         # Get the number of games with a score
#         games_with_score = len(df.dropna(subset=['teamScore', 'opponentScore']))

#         # Get the number of games without a score
#         games_without_score = total_games - games_with_score

#         # Print the results
#         print(f'Total games: {total_games}\nGames with score: {games_with_score}\nGames without score: {games_without_score}\nProportion of games with score: {games_with_score / total_games}')

#         # Group the data by team name
#         grouped = df.groupby('teamName')

#         # Calculate the summary statistics
#         summary_stats = grouped.agg({
#             'teamScore': ['count', 'mean', 'median', 'sum'],
#             'opponentScore': ['count', 'mean', 'median', 'sum']
#         }).reset_index()

#         # Define a function to calculate the results
#         def calculate_results(row):
#             if row['teamScore'] > row['opponentScore']:
#                 return 'win'
#             elif row['teamScore'] < row['opponentScore']:
#                 return 'loss'
#             else:
#                 return 'tie'

#         # Apply the function to the DataFrame
#         df['result'] = df.apply(calculate_results, axis=1)

#         # Group the data by team name and result
#         results = df.groupby(['teamName', 'result']).size().unstack(fill_value=0).reset_index()

#         # Create a DataFrame of the location summary
#         location_summary = pd.crosstab(df['teamName'], df['homeOrAway'])

#         # Create a DataFrame of the tournament summary
#         tournament_games = df[df['contestType'].notnull()]

#         tournament_summary = tournament_games.groupby('teamName').agg({
#             'teamScore': ['count', 'mean', 'median', 'sum'],
#             'opponentScore': ['count', 'mean', 'median', 'sum']
#         }).reset_index()


#         # Add a column for the year
#         summary_stats['year'] = file[-8:-4]

#         # Save the summary_stats DataFrame to a CSV file
#         summary_stats.to_csv(f'./data/processed/{file[-8:-4]}.csv', index=False)

#         # Close the function
#     return parsed_data

# # Call the function
# process_data(files)



```
