```python
## Notebook to parse game info for the ncaa tournament

## This initial run is on a strictly text document with initial teams, times ,ect
```

```python
with open('ncaa_bracket_raw.txt', 'r') as file:
    lines = file.readlines()
```

```python
import pandas as pd
import re

data = {
    "Game_Number": [],
    "Date": [],
    "Tournament": [],
    "Host": [],
    "Team_Seed": [],
    "Team_Name": [],
    "Team_Record": [],
    "Match_Time": [],
    "Network": []
}

df = pd.DataFrame(data)
game_number = 0

for line in lines:
    line = line.strip()  # Remove leading/trailing white space

    # Try to match a date line
    date_match = re.match(r"(.+), (.+), (\d+)", line)
    if date_match:
        date = f"{date_match.group(1)} {date_match.group(2)} {date_match.group(3)}"
        continue

    # Try to match a tournament line
    tournament_match = re.match(r"(.+) hosted by (.+)", line)
    if tournament_match:
        tournament = tournament_match.group(1)
        host = tournament_match.group(2)
        continue

    # Try to match a game line
    game_match = re.match(r"#(\d+) (.+) \((.+)\) vs\. #(\d+) (.+) \((.+)\), (.+), (.+)", line)
    if game_match:
        game_number += 1  # Increment the game number
        df = df.append({
            "Game_Number": game_number,
            "Date": date,
            "Tournament": tournament,
            "Host": host,
            "Team_Seed": game_match.group(1),
            "Team_Name": game_match.group(2),
            "Team_Record": game_match.group(3),
            "Match_Time": game_match.group(7),
            "Network": game_match.group(8),
        }, ignore_index=True)

        # Notice the game number is not incremented here
        df = df.append({
            "Game_Number": game_number,
            "Date": date,
            "Tournament": tournament,
            "Host": host,
            "Team_Seed": game_match.group(4),
            "Team_Name": game_match.group(5),
            "Team_Record": game_match.group(6),
            "Match_Time": game_match.group(7),
            "Network": game_match.group(8),
        }, ignore_index=True)

```
