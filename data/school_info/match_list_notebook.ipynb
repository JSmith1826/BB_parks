```python
import pandas as pd
from fuzzywuzzy import fuzz
from fuzzywuzzy import process
import numpy as np
import re
import string
```

## Notebook to get list of fields I haven't mapped yet

```python
### 

## List of all High Schools that competed in the 2022 MHSAA State Championship
teams_df = pd.read_csv('data/school_info/all_hs_list.csv')
# rename columns
teams_df = teams_df.rename(columns={'0': 'school'})

team_list = teams_df['school'].tolist()

team_list
```

```python
# Get list of HS fields that have already been plotted

plot_df = pd.read_csv('data/viz/viz_book_area.csv')
hs_list_df = plot_df[plot_df['level_x'] == 'high_school']
# hs_list_df.info()

hs_list = hs_list_df['field'].tolist()
hs_list
```

```python
### Use fuzzy match to find the closest match to from the plot list to the team list

# Create a list of tuples containing the match, the match score, and the index of the team
matches = []
for team in hs_list:
    match = process.extractOne(team, team_list)
    matches.append((match[0], match[1], hs_list.index(team)))

# Create a dataframe from the list of tuples
matches_df = pd.DataFrame(matches, columns=['match', 'score', 'index'])


## If a match is found remove it from the team list

# Create a list of the indices of the matches
match_indices = matches_df['index'].tolist()

# Create a list of the indices of the teams that were not matched
no_match_indices = [i for i in range(len(team_list)) if i not in match_indices]

# Create a list of the teams that were not matched
no_match_list = [team_list[i] for i in no_match_indices]

# Sort the dataframe by score in descending order
# matches_df = matches_df.sort_values(by='score', ascending=False)

# matches_df.info()
# matches_df.head()

len(no_match_list)

# Output no match list to csv
no_match_df = pd.DataFrame(no_match_list, columns=['school'])
no_match_df.to_csv('data/school_info/fields_to_plot_{}.csv'.format(pd.datetime.now().strftime("%Y-%m-%d %H%M%S")))


```
