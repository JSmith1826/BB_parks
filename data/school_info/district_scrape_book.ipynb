```python
#### Notebook to parse the school names of every team that played in the 2022 MHSAA Baseball State Tournament


### Links to full tourny results from MHSAA website

div_1 = 'https://my.mhsaa.com/Sports/MHSAA-Tournament-Brackets/BracketGroup/1/Classification/1/SportSeasonId/421324'
div_2 = 'https://my.mhsaa.com/Sports/MHSAA-Tournament-Brackets/BracketGroup/1/Classification/2/SportSeasonId/421324'
div_3 = 'https://my.mhsaa.com/Sports/MHSAA-Tournament-Brackets/BracketGroup/1/Classification/3/SportSeasonId/421324'
div_4 = 'https://my.mhsaa.com/Sports/MHSAA-Tournament-Brackets/BracketGroup/1/Classification/4/SportSeasonId/421324'

### Import libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re


### Get the html from the MHSAA website
div_1_html = requests.get(div_1).text
div_2_html = requests.get(div_2).text
div_3_html = requests.get(div_3).text
div_4_html = requests.get(div_4).text


### Parse the html with BeautifulSoup
div_1_soup = BeautifulSoup(div_1_html, 'html.parser')
div_2_soup = BeautifulSoup(div_2_html, 'html.parser')
div_3_soup = BeautifulSoup(div_3_html, 'html.parser')
div_4_soup = BeautifulSoup(div_4_html, 'html.parser')


```

12-15-22 Above - import cell works

```python
## Get all the team names from the MHSAA website

# tost2 = div_2_soup.find_all('li', class_='scoreItem')

tost = div_1_soup.find_all('li', class_='scoreItem')

# print(list)


# x = 140

# print(tost[x].text.split(',')[0])
# print(tost[x].text.split(',')[1])


### Extract the 2 team names from each game
team_names = []
team1 = []
team2 = []

for i in range(len(tost)):
    team1 = (tost[i].text.split(',')[0])

    team2 = (tost[i].text.split(',')[1])

# Remove last 2 characters from each team name
    team1 = team1[:-2]
    team2 = team2[:-2]

# Add the team names to the list
    team_names.append(team1)
    team_names.append(team2)


### Remove leading and trailing spaces from each team name
for i in range(len(team_names)):
    team_names[i] = team_names[i].strip()

## Add 'HS' to the end of each team name
for i in range(len(team_names)):
    team_names[i] = team_names[i] + ' HS'




#### Remove duplicates from the list
team_names = list(dict.fromkeys(team_names))



d1_df = pd.DataFrame(team_names)
# d1_df.to_csv('data\school_info\d1_district_list.csv')

# ## check list
# print(team_names)
```
